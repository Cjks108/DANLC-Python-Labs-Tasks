1. Calculate the total revenue generated by two product categories in a store
Input:
category1_revenue = np.array([500, 600, 700, 550])
category2_revenue = np.array([450, 700, 800, 600])


```python
import numpy as np

category1_revenue = np.array([500, 600, 700, 550])
category2_revenue = np.array([450, 700, 800, 600])

total_revenue = category1_revenue + category2_revenue
overall_revenue = np.sum(total_revenue)

print("Revenue per day:", total_revenue)
print("Total revenue:", overall_revenue)
```

2. Calculate the profit made by a company
Input:
revenue = np.array([10000, 12000, 11000, 10500])
expenses = np.array([4000, 5000, 4500, 4800])

```python
import numpy as np

revenue = np.array([10000, 12000, 11000, 10500])
expenses = np.array([4000, 5000, 4500, 4800])

profit = revenue - expenses
total_profit = np.sum(profit)

print("Profit per quarter:", profit)
print("Total profit:", total_profit)
```

3. Determine which products in a store are out of stock (quantity is 0).
Input:
inventory = np.array([10, 0, 5, 0, 20, 0])

```python
import numpy as np

inventory = np.array([10, 0, 5, 0, 20, 0])

out_of_stock_indices = np.where(inventory == 0)[0]

print("Out of stock product indices:", out_of_stock_indices)
```

4.Calculate the total cost of items in a shopping cart, considering the quantity and price
per item.
Input:
quantity = np.array([2, 3, 4, 1])
price_per_item = np.array([10.0, 5.0, 8.0, 12.0])

```python
import numpy as np

quantity = np.array([2, 3, 4, 1])
price_per_item = np.array([10.0, 5.0, 8.0, 12.0])

item_costs = quantity * price_per_item
total_cost = np.sum(item_costs)

print("Cost per item:", item_costs)
print("Total cost:", total_cost)
```
